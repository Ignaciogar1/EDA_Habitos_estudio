#Análisis Exploratorio de Datos (EDA) con Seaborn
##Dataset: Hábitos de estudio vs rendimiento académico

```python
#Librerías principales

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Config inicial

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

```python
#Cargar Dataset y primer vistazo

df = pd.read_csv('student_habits_performance.csv')
df.head()
```

```python
df.info()
```

```python
#Revisión de Datos Faltantes y Duplicados

print(df.isnull().sum())
print(df.duplicated().sum())
```

```python
#Rellenamos los nulos con la moda de esa columna

df['parental_education_level'] = df['parental_education_level'].fillna(df['parental_education_level'].mode()[0])
```

```python
#Convertimos variables categóricas
#Eliminamos columna student_id

categorical_cols = ['gender', 'part_time_job', 'diet_quality', 
                    'parental_education_level', 'internet_quality', 
                    'extracurricular_participation']

for col in categorical_cols:
    df[col] = df[col].astype('category')
```

```python
#Graficamos las variables Numéricas

numeric_cols = ['age','study_hours_per_day','social_media_hours', 'netflix_hours', 
                'attendance_percentage', 'sleep_hours', 'exercise_frequency', 
                'mental_health_rating','exam_score']

col_names_es = {
    'age': 'Edad',
    'study_hours_per_day': 'Horas de estudio/día',
    'social_media_hours': 'Horas en redes sociales',
    'netflix_hours': 'Horas en Netflix',
    'attendance_percentage': 'Porcentaje de asistencia',
    'sleep_hours': 'Horas de sueño',
    'exercise_frequency': 'Frecuencia de ejercicio',
    'mental_health_rating': 'Salud mental',
    'exam_score': 'Puntuación examen'
}

paleta = sns.color_palette("pastel")
fig, axes = plt.subplots(3, 3, figsize=(10,8))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    sns.histplot(df[col], bins=10, ax=axes[i], color=paleta[i % len(paleta)], edgecolor='black')
    axes[i].set_xlabel(col_names_es[col])
    axes[i].set_ylabel('Número de estudiantes')
    axes[i].set_title('')  # quitar título para estilo más limpio

# Ocultar subplots vacíos si los hay
for j in range(len(numeric_cols), len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

```python
#Graficamos las variables Categóricas

categorical_cols = [
    'gender', 'part_time_job', 'diet_quality', 
    'parental_education_level', 'internet_quality', 'extracurricular_participation'
]

# Nombres en español
cat_names_es = {
    'gender': 'Género',
    'part_time_job': 'Trabajo a tiempo parcial',
    'diet_quality': 'Calidad de dieta',
    'parental_education_level': 'Nivel educativo de los padres',
    'internet_quality': 'Calidad de internet',
    'extracurricular_participation': 'Participación extracurricular'
}

paleta = sns.color_palette("pastel")

fig, axes = plt.subplots(2, 3, figsize=(8,6))
axes = axes.flatten()

for i, col in enumerate(categorical_cols):
    sns.countplot(x=col, data=df, ax=axes[i], color=paleta[i % len(paleta)], edgecolor='black')
    axes[i].set_xlabel(cat_names_es[col])
    axes[i].set_ylabel('Número de estudiantes')
    axes[i].set_title('')  # sin título redundante
    axes[i].tick_params(axis='x', rotation=30)  # rotar etiquetas si son largas

plt.tight_layout()
plt.show()
```

```python
#Matriz de Correlación

plt.figure(figsize=(10,8))
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de correlación')
plt.show()
```

```python
#Estadísticas descriptivas

df[categorical_cols].describe()  # para categóricas
df.describe()  # para numéricas
```

```python
# Promedio y desviación de examen por nivel de dieta
df.groupby('diet_quality', observed=True)['exam_score'].agg(['mean','std'])

# Promedio y desviación por nivel educativo de los padres
df.groupby('parental_education_level', observed=True)['exam_score'].agg(['mean','std'])

# Promedio y desviación por calidad de internet
df.groupby('internet_quality', observed=True)['exam_score'].agg(['mean','std'])
```
